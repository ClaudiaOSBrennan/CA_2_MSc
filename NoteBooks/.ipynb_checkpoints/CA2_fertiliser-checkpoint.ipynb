```python
df_fertiliser.drop(columns=["STATISTIC","Statistic","C02069V02500",
                            "UNIT","TLIST(A1)"],inplace=True)
```

```python
df_fertiliser.dtypes
```

```python
df_fertiliser["VALUE"].astype("float")
```

```python
df_fertiliser = df_fertiliser[~(df_fertiliser["VALUE"] == ".")].copy()
```

```python
df_fertiliser["VALUE"] = df_fertiliser["VALUE"].astype("float")
```

```python
df_fertiliser
```

```python
df_fertiliser = df_fertiliser.pivot_table(values="VALUE", columns="Type of Fertiliser", index="Year").reset_index()
```

```python
df_fertiliser
```

```python
df_fertiliser.columns.name=""
```

```python
df_fertiliser.isna().sum()
```

```python
columns = []
for column in df_fertiliser.isna().sum()[df_fertiliser.isna().sum()<5].index:
    columns.append(column)
columns
```

```python
df_fertiliser=df_fertiliser[columns]
```

```python
sns.heatmap(data=df_fertiliser.corr(),annot=True)
```

```python
df_fertiliser[["Compound 10-10-20","Compound 14-7-14"]].head(3)
```

```python
sns.scatterplot(data=df_fertiliser, x="Compound 10-10-20", y="Compound 14-7-14")
```

```python
plt.rcParams["figure.figsize"] = (12,6)
df_fertiliser.groupby("Year").mean().plot()
```

```python
sns.scatterplot(data = df_fertiliser, x="Calcium Ammonium Nitrate (27.5% N)",
            y="Compound 0-10-20")
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(df_fertiliser.drop(columns=["Year"]).dropna())
print(pca.explained_variance_ratio_)
```

```python
df_fertiliser.isna().sum()
```

```python
pca = PCA(n_components=1)
pca.fit(df_fertiliser.drop(columns=["Year","Muriate of Potash (50% K)"]).dropna())
print(pca.explained_variance_ratio_)
```

```python
df_fertiliser["fertiliser_pca"] = pca.transform(df_fertiliser.drop(columns=["Year","Muriate of Potash (50% K)"]).dropna())
```

```python
df_fertiliser = df_fertiliser[["Year","fertiliser_pca"]]
```
